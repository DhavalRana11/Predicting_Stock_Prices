```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
```

```python
df = pd.read_csv("TSLA.csv")
df
```

```python
df = df.rename(columns={'Date':'date','Open':'open','High':'high','Low':'low','Close':'close','Adj Close':'adj close','Volume':'Volume'})
df.head()
```

```python
df['date']= pd.to_datetime(df.date)
df
```

```python
fig = px.line(df, x='date', y='close', title='tesla close price')
fig.show()
```

```python
closedf = df[['date', 'close']]
closedf
```

```python
training_size = int (len(closedf)*0.70)
test_size = len(closedf)-training_size
train_data, test_data= closedf[0:training_size,:], closedf[training_size:len(closedf),:1]
```

```python
test_data
```

```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 20 
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)

print('x_train', x_train.shape)
print('y_train', y_train.shape)
print('x_test', x_test.shape)
print('y_test', y_test.shape)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 1000)
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)

print(pred)
```

```python
print(y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
```

```python
print("mean absolute error=" + str(mean_absolute_error(y_test, pred)))
print("Root mean Squared error - MSE ="+ str(math.sqrt(mean_squared_error(y_test, pred))))
```
